```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.uniform(0,1,1000) #uses np.random.uniform() to get intergers between 2 numbers instead of random.randint() to get int between 2 ints
```

```python
def hist(x): #defines hist function
    
    save_flag = True #allows figure to save be saved depending on what the user wants 
    xlabel = "x" #labels x axis 
    ylabel = "y    " #Labels y axis
    lcolor = "purple" #defines the color that the lines are filled in by
    fs = 14
    fname = "histplot.pdf"
    
    fig, ax = plt.subplots(1,1, figsize=(12,4)) # the 1,1 defines how many plots are on the row, column. Figsize defines the x scale, y scale

    ax.hist(x,color = lcolor, bins = 100, edgecolor = "black", alpha = 0.5) #ax.hist make the graph be a histogram, with everything else changing attributes of the histogram

    ax.set_xlabel(xlabel,fontsize = fs)
    ax.set_ylabel(ylabel,fontsize = fs, rotation = 0) #rotation defines how much the label is rotated naturally
    if(save_flag == True):
        plt.savefig(fname, bbox_inches="tight", dpi = 400)
```

```python
hist(x) #plots historgram by calling function
```
