```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = []
for i in range(101):
    x.append(i)
```

```python
def co(x):
    return np.cos(x)
    
def si(x):
    return np.sin(x)
```

```python
def multiplot(x, y1, y2, flag_save = True, xlabel = "x", fs = 14, fname = "plot.pdf"):  #defines plot parameters 
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(11,4)) #defines the number of plots (ax2 and ax2) also defines the figure size
    
    ax1.plot(x,si(x),linewidth=1.5)
    ax1.set_xlabel(xlabel,fontsize=fs)
    ax1.set_ylabel(ylabel = "sin(x)",fontsize=fs)
    
    ax2.plot(x,co(x),linewidth=1.5) #plots x vs cos(x)
    ax2.set_xlabel(xlabel,fontsize=fs) #sets x axis to parameter set earlier
    ax2.set_ylabel(ylabel = "cos(x)",fontsize=fs) #sets y axis to manually set name
    
    if(flag_save):
        plt.savefig(fname,bbox_inches='tight',dpi=400)

    plt.tight_layout(pad=4.0)  #seperates figure 

multiplot(x,co(x),si(x))  #inputs values into parameters defined earlier to graph function
```
