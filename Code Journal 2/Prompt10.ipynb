```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(1,41,1000)
```

```python
def hist(x):
    save = True
    xlabel = "x"
    ylabel = "y    "
    lcolor = "orange"
    fs = 14
    fname = "hist.pdf"

    fig,ax = plt.subplots(1,1, figsize = (12,4))

    ax.hist(x, color=lcolor, bins = 100, edgecolor = "black", alpha = 1)

    ax.set_xlabel(xlabel, fontsize = fs)
    ax.set_ylabel(ylabel, fontsize = fs, rotation = 0)
    
    if(save == True):
        plt.savefig(fname, bbox_inches = "tight", dpi=400)
```

```python
hist(x)
```
