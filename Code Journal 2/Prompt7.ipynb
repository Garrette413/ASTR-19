```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = []
add = 0
for i in range(100):
    x.append(round(add,3))
    add = add + (1/99)
```

```python
def e(x):
    return np.exp(x)
```

```python
def plot(x,y, flag_save = True, xlabel = "Time [milliseconds]", ylabel = "Awesomeness", fs = 14, fname = "plot.pdf"):
    
    fig, ax = plt.subplots(1,1,figsize=(4,4))
    ax.plot(x,y,linewidth=1.5)
    ax.set_xlabel(xlabel,fontsize=fs)
    ax.set_ylabel(ylabel,fontsize=fs)
    if(flag_save):
        plt.savefig(fname,bbox_inches='tight',dpi=400)
```

```python
y = e(x)
```

```python
plot(x,y)
```
