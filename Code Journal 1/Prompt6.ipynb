```python
import math as m
```

```python
def co(x):
    return m.cos(x)
```

```python
def si(x):
    return m.sin(x)
```

```python
x = 2*m.pi
```

```python
cos = []
sin = []
for i in range(1000):
    cos.append(co(x-(i*x/1000)))
    sin.append(si(x-(i*x/1000)))
```

```python
print(" |        sin(x)       |        cos(x)      |")
for i in range(1000):    
    print(" |      %-6.5f       |      %-6.5f       |" % (sin[i], cos[i]))
```

```python
print(" |        sin(x)       |        cos(x)      |")
for i in range(10):    
    print(" |      %-6.5f       |      %-6.5f       |" % (sin[i], cos[i]))
```
